# Пример 4.3.1

```python
from statistics import mean, median
import random

import numpy as np
import matplotlib.pyplot as plt

n = random.randint(10, 50)
items = np.random.rand(n)
#print(items)
valMedian = median(items) # берет средние значение из массива
valMiddle = mean(items) # среднее арифметическое значений элементов массива

print(valMedian, valMiddle)

fig, ax = plt.subplots()
x = items

ax.scatter(x, items, c = 'blue')    #  цвет точек

ax.set_facecolor('white')
ax.set_title('Точечную диаграмма рассения ряда')

fig.set_figwidth(8)     #  ширина и
fig.set_figheight(8)    #  высота "Figure"

plt.show()
```

# Пример 4.3.2

```python
import math

def expression(x):
      a = math.sqrt(1 + math.exp(math.sqrt(x)) + (math.cos(x)**2))
      b = abs(1 - math.sin(x)**3)
      c = math.log10(2*x)
      return a/b+c

arr = [expression(random.randint(1, 10)) for x in range(10)]
arr1 = arr[:5:]
plt.title("линейный график")
plt.plot(arr, arr, color='red')
plt.show()

plt.title("точечный график")
plt.plot(arr1, arr1, 'ro')
plt.show()
```

# Пример 4.3.5

```python
import math as m
print("Ноль в качестве знака операции\nзавершает работу программы")
while True:
      s = input("Знак (+,-,*,/, sin, cos, e^(x+y), x^y): ")
      if s == '0':
            break
      if s in ('+', '-', '*', '/', 'sin', 'cos', 'e^(x+y)', 'x^y'):
            x = float(input("x="))
            y = float(input("y="))
            if s == 'sin':
                  print(m.sin(x+y))
            elif s == 'cos':
                  print(m.cos(x+y))
            elif s == 'e^(x+y)':
                  print(m.exp(x+y))
            elif s == 'x^y':
                  print(m.pow(x,y))
            elif s == '+':
                  print("%.2f" % (x+y))
            elif s == '-':
                  print("%.2f" % (x-y))
            elif s == '*':
                  print("%.2f" % (x*y))
            elif s == '/':
                  if y != 0:
                        print("%.2f" % (x/y))
                  else:
                        print("Деление на ноль!")
      else:
           print("Неверный знак операции!")
```
